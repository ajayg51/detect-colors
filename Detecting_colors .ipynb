```python
import cv2
import numpy as np
```

```python
def empty(a):
    pass

cv2.namedWindow("HSV")
cv2.resizeWindow("HSV",680,900)
cv2.moveWindow("HSV",500,500)
cv2.createTrackbar("Hue-Min","HSV",0,179,empty)
cv2.createTrackbar("Hue-Max","HSV",179,179,empty)
cv2.createTrackbar("Sat-Min","HSV",0,255,empty)
cv2.createTrackbar("Sat-Max","HSV",255,255,empty)
cv2.createTrackbar("Val-Min","HSV",0,255,empty)
cv2.createTrackbar("Val-Max","HSV",255,255,empty)
cv2.moveWindow("HSV",600,0)
```

```python
cap = cv2.VideoCapture(0)
while True:
#         Getting values of trackBars
        hmin=cv2.getTrackbarPos("Hue-Min","HSV")
        hmax=cv2.getTrackbarPos("Hue-Max","HSV")
        smin=cv2.getTrackbarPos("Sat-Min","HSV")
        smax=cv2.getTrackbarPos("Sat-Max","HSV")
        vmin=cv2.getTrackbarPos("Val-Min","HSV")
        vmax=cv2.getTrackbarPos("Val-Max","HSV")
        lower=np.array([hmin,smin,vmin])
        upper=np.array([hmax,smax,vmax])
        
    
        ret,frame = cap.read()
    #     RGB to HSV
        hsv_frame=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
        
    #     upper and lower limit of RGB , creating respective using these values
        low_red=np.array([161,155,84])
        high_red=np.array([179,255,255])
        red_mask=cv2.inRange(hsv_frame,low_red,high_red)
        red=cv2.bitwise_and(frame,frame,mask=red_mask)
        low_green=np.array([25,52,72])
        high_green=np.array([102,255,255])
        green_mask=cv2.inRange(hsv_frame,low_green,high_green)
        green=cv2.bitwise_and(frame,frame,mask=green_mask)
        low_blue=np.array([94,80,2])
        high_blue=np.array([126,255,255])
        blue_mask=cv2.inRange(hsv_frame,low_blue,high_blue)
        blue=cv2.bitwise_and(frame,frame,mask=blue_mask)
    #     Every color except White
        low=np.array([0,42,0])
        high=np.array([179,255,255])
        mask=cv2.inRange(hsv_frame,low,high)
        Color=cv2.bitwise_and(frame,frame,mask=mask)
    #     Just White
        low_white=np.array([0,0,168])
        high_white=np.array([172,111,255])
        white_mask=cv2.inRange(hsv_frame,low_white,high_white)
        white=cv2.bitwise_and(frame,frame,mask=white_mask)
        f=cv2.resize(frame,(400, 400))
        

        #     Change values of Track bar and get different shades of video
        
        diffShades_mask=cv2.inRange(hsv_frame,lower,upper)
        diffShades=cv2.bitwise_and(frame,frame,mask=diffShades_mask)
        frame="FRAME"
    #   Moving window to topmost position giving y value of -100 and changing values of y of other windows accordingly."
        
        x=0
        y=-100
        width=400
        height=400
        cv2.moveWindow(frame,x,y)
        cv2.imshow(frame,f)
        r=cv2.resize(red,(width,height))
        red="RED"
        cv2.moveWindow(red,x+width+1,y)
        cv2.imshow(red,r)
        g=cv2.resize(green,(width, height))
        green="GREEN"
        cv2.moveWindow(green,x+2*width+1,y)
        cv2.imshow(green,g)
        b=cv2.resize(blue,(width,height))
        blue="BLUE"
        cv2.moveWindow(blue,x,height+y+1)
        cv2.imshow(blue,b)
        nW=cv2.resize(Color,(width,height))
        Color="COLORs EXCEPT WHITE"
        cv2.moveWindow(Color,x+width+1,height+y+1)
        cv2.imshow(Color,nW)
        w=cv2.resize(white,(width,height))
        white="WHITE"
        cv2.moveWindow(white,x+1+2*width,height+y+1)
        cv2.imshow(white,w)
        diffShades=cv2.resize(diffShades,(500,500))
        cv2.imshow("Different Shades",diffShades)
        key=cv2.waitKey(1)
        if key==27:
            break
    # When everything done, release the capture\n",
cap.release()
cv2.destroyAllWindows() 
   
```
